##Trabajo Practico - Ciencia de datos y datos masivos

### Creacion de base de datos

```python
%sql
CREATE DATABASE IF NOT EXISTS Vehicle_Storage;
USE Vehicle_Storage;
```

## Creacion de tablas

#### Tabla marcas

```python
%sql
create table marcas (
    id_marca int,
    nombre_marca string,
    pais_marca string,
    establishment_year int
)
```

#### Tabla vehiculos

```python
%sql
create table vehiculos (
    id_vehiculo int,
    id_marca int,
    modelo string,
    anio int,
    precio double,
    kilometraje int,
    estado string,       --Es el estado de venta (disponible, vendido, reservado)
    fecha_creacion date

)
```

#### Tabla clientes

```python
%sql
create table customers (
    id_customer int,
    nombre_customer string,
    email string,
    telefono string,
    direccion string,
    customer_creation date
)
```

#### Tabla pagos

```python
%sql
create table ordenes (
    id_orden int,
    id_customer int,
    id_vehiculo int,
    tipo_orden string,  -- Si es compra o reserva
    status_order string, -- Pendiente, completada,cancelada
    order_date date
)
```

#### Tabla pagos

```python
%sql
create table pagos (
    id_pago int,
    id_orden int,
    monto double,
    date_pago date,
    metodo_pago string,  --Transferencia, contado, financiacion
    status_pago string -- Pendiente,pagado
)
```

##Consultas

#### Consulta 1 - Autos disponibles y marcas

```python
%sql
SELECT v.modelo, m.nombre_marca
FROM vehiculos v
JOIN marcas m ON v.id_marca = m.id_marca
WHERE v.estado = 'disponible';
```

#### Consulta 2 - Total recaudado

```python
%sql
SELECT metodo_pago, SUM(monto) AS total_recaudado
FROM pagos
WHERE status_pago = 'pagado'
GROUP BY metodo_pago;
```

#### Consulta 3 - Clientes con ordenes completadas

```python
%sql
SELECT c.nombre_customer, v.modelo, v.anio, m.nombre_marca
FROM ordenes o
JOIN customers c ON o.id_customer = c.id_customer
JOIN vehiculos v ON o.id_vehiculo = v.id_vehiculo
JOIN marcas m ON v.id_marca = m.id_marca
WHERE o.status_order = 'completada';
```

#### Consulta 4 - Ranking de marcas por ventas

```python
%sql
SELECT m.nombre_marca, COUNT(*) AS total_ventas
FROM ordenes o
JOIN vehiculos v ON o.id_vehiculo = v.id_vehiculo
JOIN marcas m ON v.id_marca = m.id_marca
WHERE o.status_order = 'completada'
GROUP BY m.nombre_marca
ORDER BY total_ventas DESC;
```

#### Consulta 5 - Autos que no fueron ordenados

```python
%sql
SELECT v.modelo, m.nombre_marca
FROM vehiculos v
JOIN marcas m ON v.id_marca = m.id_marca
LEFT JOIN ordenes o ON v.id_vehiculo = o.id_vehiculo
WHERE o.id_vehiculo IS NULL;
```

- Una orden corresponde a un solo auto  
- Un pago corresponde a una orden
